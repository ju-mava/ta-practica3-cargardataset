# Creando un dataset para HuggingFace

En este notebook veremos cómo crear un dataset.

Para este notebook la única librería que vamos a necesitar es la de [datasets](https://huggingface.co/course/chapter5/5).

```python
!pip install datasets -Uq
```

Ahora ya podemos crear nuestro dataset. Según aparece en la [documentación de HuggingFace](https://huggingface.co/course/chapter5/2), podemos construir un dataset a partir de un fichero csv, un json, un fichero de texto plano o un fichero de pickle. Aunque aparezca la opción de construir un dataset a partir de un fichero de texto plano, este debe tener una estructura determinada y es por eso por lo que hemos construido nuestro csv.

Pasamos a cargar el dataset a partir de dicho fichero csv.

```python
from datasets import load_dataset
```

```python
es_en_dataset = load_dataset("csv", data_files="caperucita_roja.csv",delimiter=";")
es_en_dataset
```

Si nos fijamos, el dataset que se ha cargado solo tiene un conjunto de entrenamiento, y como hemos comentado con anterioridad, siempre interesa tener un conjunto de test además de uno de entrenamiento. Vamos a fijar dichos conjuntos para que cualquiera que use nuestro dataset use siempre las mismas particiones.

```python
es_en_dataset = es_en_dataset["train"].train_test_split(test_size=0.2)
es_en_dataset
```

Ya hemos creado nuestro dataset y está listo para subirlo a HuggingFace. Para ello nos deberemos logear usando el token de escritura que tenemos de la asignatura de Aprendizaje Automático.

```python
from huggingface_hub import notebook_login

notebook_login()
```

Ahora ya podemos subir nuestro dataset con el método `push_to_hub` indicando el nombre del dataset.

```python
es_en_dataset.push_to_hub("es_en")
```

Una vez subido el dataset este estará disponible en la siguiente página de HuggingFace: https://huggingface.co/datasets/joheras/adv-ele (con el nombre de usuario correspondiente). Es conveniente que una vez se haya subido el dataset, se añada una [dataset card](https://github.com/huggingface/datasets/blob/main/CONTRIBUTING.md#how-to-contribute-to-the-dataset-cards) con información detallada sobre el mismo.

Finalmente, podemos descargar nuestro dataset como cualquier otro dataset disponible en HuggingFace.

```python
remote_dataset = load_dataset("jumava/es_en")
remote_dataset
```
